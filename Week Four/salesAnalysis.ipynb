```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("sales_dataset.csv")
```

```python
df.columns
```

```python
city = df.groupby("city")["total_amount"].sum()*100/df["total_amount"].sum()
print(city)
```

```python
label = ["Bangalore","Delhi","Hyderabad","Mumbai","Nagpur","Pune"]
plt.pie(city,labels=label,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Distribution of Total Sales by City")
plt.show()
```

```python
product = df.groupby("product")["total_amount"].sum()*100/df["total_amount"].sum()
print(product)
```

```python
label = ["Headphones","Keyboard","Laptop","Mobile","Mouse","Smartwatch"]
plt.pie(product,labels=label,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Distribution of Total Sales by Product")
plt.show()
```

```python
quantity = df.groupby("product")["quantity"].sum()
print(quantity.index)
```

```python
plt.figure(figsize=(10,10))
bars = plt.bar(quantity.index,quantity.values)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 10, int(yval), ha='center', va='bottom')

plt.title("Total Quantity of Product Sold")
plt.xlabel("Products")
plt.ylabel("Quantity Sold")
```
